```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langgraph.checkpoint.memory import MemorySaver
```

```python
# Define state
class ChatState(TypedDict):
    message: str
    response: str
```

```python
# Node
def chatbot(state: ChatState):
    user_msg = state["message"]
    return {"response": f"Bot reply to: {user_msg}"}
```

```python
# Create graph
graph = StateGraph(ChatState)
graph.add_node("chatbot", chatbot)

graph.add_edge(START, "chatbot")
graph.add_edge("chatbot", END)
```

```python
# Add persistence (memory)
memory = MemorySaver()
```

```python
# Compile with persistence
workflow = graph.compile(checkpointer=memory)
```

```python
# Run with thread id (important for persistence)
config = {"configurable": {"thread_id": "user1"}}
```

```python
# First interaction
result1 = workflow.invoke({"message": "Hello"}, config=config)
print(result1)
```

```python
# Second interaction (same thread)
result2 = workflow.invoke({"message": "How are you?"}, config=config)
print(result2)
```

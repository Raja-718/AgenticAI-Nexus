```python
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI # 1. Import Chat version
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
# 2. Use the correct model ID and class
model = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.7)
```

```python
class LLMState(TypedDict):
    question: str
    answer: str
```

```python
def llm_qa(state: LLMState) -> LLMState:
    question = state['question']
    prompt = f"Answer the following question: {question}"
    
    # 3. .content is REQUIRED here because it's a Chat model
    response = model.invoke(prompt)
    answer_text = response.content 

    return {"question": question, "answer": answer_text}
```

```python
# Define Graph
graph = StateGraph(LLMState)
graph.add_node('llm_qa', llm_qa)
graph.add_edge(START, 'llm_qa')
graph.add_edge('llm_qa', END)

workFlow = graph.compile()
```

```python
# Execute
initial_state = {"question": "how far the moon is from earth?"}
final_state = workFlow.invoke(initial_state)
print(final_state['answer'])
```
